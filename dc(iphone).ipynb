```
import pandas as pd
```

```
df=pd.read_csv("iphone_sales_dataset.csv")
```

```
print(df.head())
```

```
print(df.info())
```

```
print(df.describe())
```

```
print(df.isnull().sum())
```

```
df=df.dropna()
```

```
print(df.duplicated().sum())
```

```
df=df.drop_duplicates()
```

```
print(df.dtypes)
```

```
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df['Price'] >= lower) & (df['Price'] <= upper)]
```
